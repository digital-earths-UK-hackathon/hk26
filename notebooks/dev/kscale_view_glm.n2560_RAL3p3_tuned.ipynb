```python
from pathlib import Path

import cartopy.crs as ccrs
import iris
import matplotlib.pyplot as plt
import xarray as xr
```

```python
basedir = Path('/gws/nopw/j04/kscale/DYAMOND3_reruns/5km-RAL3p3-tuned/glm/field.pp')
```

```python
streams = sorted(basedir.glob('apver?.pp'))
streams
```

```python
paths_a = sorted(streams[0].glob('glm.n2560_RAL3p3_tuned.*.pp'))
paths_a[:10]
```

```python
# Have to use iris to load UM .pp output.
cubes = iris.load(paths_a[0])
cubes
```

```python
# Convert to xr.DataArray (easier to work with)
da = xr.DataArray.from_iris(cubes.extract_cube('air_temperature'))
```

```python
da
```

```python
# quickplot
da.isel(time=1).plot()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})
ax.pcolormesh(da.longitude, da.latitude, da.isel(time=1))
ax.coastlines()
```
