# Display a Hovmöller diagram over the tropics

* healpix is a little different in terms of how you deal with coordinates.
* You've seen this if you've limited the global model to a particular domain.
* This demonstrates how to plot a Hovmöller by using the `.groupby_bins` method.

```python
import intake
import numpy as np

from utils import hp_mods
```

```python
# Filter out annoying warning.
import warnings
warnings.filterwarnings("ignore", message=".*The return type of `Dataset.dims` will be changed.*", category=FutureWarning)
```

```python
# Open catalog.
url = 'https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml'
cat = intake.open_catalog(url)['UK']
# Use online if not on JASMIN.
# cat = intake.open_catalog('https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml')['online']
```

```python
sim = 'um_glm_n2560_RAL3p3_tuned_hk26'
sim_cat = cat[sim]
```

```python
ds1h = sim_cat(zoom=8, time='PT1H').to_dask().pipe(hp_mods)
```

```python
# Select 1 month, and -15 < lat < 15.
pr_eq = ds1h.pr.isel(time=slice(24 * 30), cell=(ds1h.lat > -15) & (ds1h.lat < 15))
# Groupy lon by bins (mean within each bin) -- retains time axis.
hov_pr = pr_eq.groupby_bins('lon', np.linspace(45, 180, 136)).mean()

hov_pr.plot()
```
