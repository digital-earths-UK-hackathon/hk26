# 2026 Global and regional hackathon dataset URLs

* For the 2026 hackathon: https://digital-earths-uk-hackathon.github.io/

## Open dataset from catalog

```python
from itertools import product

import intake
import pandas as pd
```

```python
assert intake.__version__ == '0.7.0', 'intake version 0.7.0 required'
```

```python
# Open catalog and print hk26 available models.
url = 'https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml'
cat = intake.open_catalog(url)['online']
# Use online if not on JASMIN.
# cat = intake.open_catalog('https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml')['online']
print('\n'.join([k for k in cat if k.endswith('_hk26')]))
```

```python
# Generate all URLs.
data = []
freqs = ['PT1H', 'PT3H']
zooms = range(0, 11)[::-1]
for sim, freq, zoom in product([k for k in cat if k.endswith('_hk26')], freqs, zooms):
    try:
        url = cat[sim](zoom=zoom, freq=freq).urlpath[0]
    except ValueError:
        # Not all zooms available for all sims.
        continue
    data.append((sim, zoom, freq, url))

df = pd.DataFrame(columns=['sim', 'zoom', 'freq', 'url'], data=data)
```

```python
print(df[df.sim == 'um_glm_n2560_RAL3p3_tuned_hk26'].to_markdown())
```
